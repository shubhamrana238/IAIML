```python
df['ApplicantIncome'].plot()
```

A line plot is a type of graph that shows how something changes over time or another continuous period.

```python
df.plot.scatter(x='ApplicantIncome', y='CoapplicantIncome')
```

A scatter plot is a way to show individual points on a graph. Each point has a position on the graph, determined by its values on two different variables. It's like marking points on a coordinate grid. Scatter plots are often used to see if there's any relationship or pattern between two sets of data.

```python
sns.barplot(x='Loan_Amount_Term',y='Credit_History',data=df)
```

A bar plot is a type of graph that uses rectangular bars to show and compare different categories of data. Each bar's height represents the quantity or value associated with that category.

```python
sns.countplot(x="Dependents",data=df)
```

A count plot is a type of graph that counts how many times something appears in a dataset and then shows that count using bars.

```python
df['Loan_Amount_Term'].plot.hist()
```

A histogram is a type of graph that groups data into "buckets" or "bins" and shows how many data points fall into each bucket.

```python
sns.violinplot(y=df["Loan_Amount_Term"])
```

A violin plot is a type of graph that combines a box plot and a density plot. It's used to display the distribution of data across different categories or groups.

```python
sns.pairplot(df,hue="Education")
```

A pair plot is a type of graph that helps you compare pairs of features or variables in a dataset. It's particularly useful when you have multiple variables and you want to understand how they relate to each other.
